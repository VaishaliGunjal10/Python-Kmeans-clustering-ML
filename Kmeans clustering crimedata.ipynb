```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fn= pd.read_csv("crime_data.csv")
fn
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
fn_new = pd.DataFrame(scaler.fit_transform(fn.iloc[:,1:]))
fn_new 
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(fn_new, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(fn_new) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
fn['clust']=cluster_labels # creating a  new column and assigning it to new column 
fn 
```

```python
fn.iloc[:,1:].groupby(fn['clust']).mean()
```

```python
fn.plot(x="Murder",y ="Rape",c=cluster_labels,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```
